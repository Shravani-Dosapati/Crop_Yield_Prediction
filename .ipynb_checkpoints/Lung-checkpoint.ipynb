```python
plt.figure(figsize = (10,8))
for i, col in enumerate (lung_data.select_dtypes(include = 'number').columns):
    plt.subplot(1,3, i+1)
    plt.boxplot(lung_data[col])
    plt.title(lung_data[col])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 10))
for i, col in enumerate(lung_data.select_dtypes(include = 'object').columns):
    plt.subplot(4, 6, i + 1)
    # Use pd.crosstab for a clean view of counts
    ct = pd.crosstab(lung_data[col], lung_data['Final_Prediction'], normalize='index') * 100
    ct.plot(kind='bar', stacked=True, ax=plt.gca())
    plt.title(f'Prediction by {col}')
    plt.ylabel('Percentage')
    plt.xticks(rotation=45, ha='right')
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
# Use the 'rocket' palette to highlight the difference
sns.barplot(x=lung_data['Final_Prediction'].index, y=lung_data['Final_Prediction'].values, palette="rocket")
plt.title('Distribution of Final_Prediction (Target Imbalance Check)')
plt.xlabel('Prediction')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Assuming df is your DataFrame

# Identify categorical columns (modify this list based on your df.info() output)
categorical_features = lung_data.columns

print("--- Chi-Squared Test of Independence (Categorical Features) ---")

for col in categorical_features:
    # 1. Create a contingency table (crosstab)
    contingency_table = pd.crosstab(lung_data[col], lung_data['Final_Prediction'])
    
    # 2. Perform the Chi-Squared Test
    # chi2: Chi-Square statistic
    # p: Two-sided p-value
    # dof: Degrees of freedom
    # expected: Expected frequencies
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print(f"\nFeature: {col}")
    print(f"P-Value: {p:.4f}")
    
    if p < 0.05:
        print("Conclusion: Significant relationship (Reject Null Hypothesis of Independence)")
    else:
        print("Conclusion: No significant relationship (Fail to Reject Null Hypothesis)")
```
